```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.decomposition import PCA, KernelPCA
from sklearn.manifold import LocallyLinearEmbedding

# 生成环形数据
X, _ = make_circles(n_samples=400, factor=.3, noise=.05)

# 可视化环形数据
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Simulated Circular Data")
plt.xlabel("X1")
plt.ylabel("X2")
plt.axis('equal')
plt.show()

# 应用 PCA、KPCA 和 LLE 算法进行降维
# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# KPCA
kpca = KernelPCA(n_components=2, kernel="rbf", gamma=15)
X_kpca = kpca.fit_transform(X)

# LLE
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_lle = lle.fit_transform(X)

# 可视化降维结果
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# PCA 可视化
axs[0].scatter(X_pca[:, 0], X_pca[:, 1], s=50)
axs[0].set_title('PCA Result')
axs[0].set_xlabel('Principal Component 1')
axs[0].set_ylabel('Principal Component 2')

# KPCA 可视化
axs[1].scatter(X_kpca[:, 0], X_kpca[:, 1], s=50)
axs[1].set_title('KPCA Result')
axs[1].set_xlabel('Kernel PC 1')
axs[1].set_ylabel('Kernel PC 2')

# LLE 可视化
axs[2].scatter(X_lle[:, 0], X_lle[:, 1], s=50)
axs[2].set_title('LLE Result')
axs[2].set_xlabel('LLE Component 1')
axs[2].set_ylabel('LLE Component 2')

plt.tight_layout()
plt.show()
```
